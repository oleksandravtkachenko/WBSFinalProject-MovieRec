```python
def build_query(mood, genres, favorite, actor=None, avoid=None):
    q = f"I enjoy {', '.join(genres)} movies. My mood is {mood}."
    if favorite:
        q += f" I loved {favorite}."
    if actor:
        q += f" I want a movie starring {actor}."
    if avoid:
        q += f" Avoid: {', '.join(avoid)}."
    return q

def build_couple_query(user_a, user_b, alpha, beta):
    return f"""
    Person A (weight {alpha:.0%}) enjoys {', '.join(user_a['genres'])},
    mood: {user_a['mood']}, loved: {user_a['favorite']}.
    Person B (weight {beta:.0%}) enjoys {', '.join(user_b['genres'])},
    mood: {user_b['mood']}, loved: {user_b['favorite']}.
    Find movies both would genuinely enjoy.
    """
```

```python
# Test 1 — basic query
query = build_query("Happy", ["Comedy", "Romance"], "Notting Hill")
print("Test 1 — Basic query:")
print(query)
print()
```

```python
# Test 2 — with actor and avoid
query = build_query("Tense", ["Thriller", "Action"], "Inception",
                    actor="Leonardo DiCaprio", avoid=["Horror", "Gore"])
print("Test 2 — With actor and avoid:")
print(query)
print()
```

```python
# Test 3 — couple query
user_a = {"genres": ["Action", "Sci-Fi"], "mood": "Adventurous", "favorite": "Interstellar"}
user_b = {"genres": ["Romance", "Drama"], "mood": "Romantic", "favorite": "Notting Hill"}

query = build_couple_query(user_a, user_b, alpha=0.52, beta=0.48)
print("Test 3 — Couple query:")
print(query)
```

```python
# Test 4 — weights sum to 1
alpha, beta = 0.52, 0.48
print(f"Test 4 — Weights sum to 1: {round(alpha + beta, 2) == 1.0}")
```
