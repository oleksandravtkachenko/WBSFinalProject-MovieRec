```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from dotenv import load_dotenv
import os

load_dotenv()

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=os.getenv("SPOTIFY_ID"),
    client_secret=os.getenv("SPOTIFY_SECRET")
))

def get_playlist(title, year):
    # Search for the specific movie soundtrack
    query = f"{title} {year} official soundtrack"
    res = sp.search(q=query, type="playlist", limit=3)
    items = res["playlists"]["items"]

    if items:
        # Try to find the most relevant result
        for item in items:
            name = item["name"].lower()
            # Prefer official soundtracks
            if any(word in name for word in ["soundtrack", "score", "ost", title.lower()]):
                return {
                    "name": item["name"],
                    "url": item["external_urls"]["spotify"],
                    "image": item["images"][0]["url"] if item["images"] else None
                }
        # Fallback to first result
        p = items[0]
        return {
            "name": p["name"],
            "url": p["external_urls"]["spotify"],
            "image": p["images"][0]["url"] if p["images"] else None
        }
    return None
```

```python
# Test 1 — Get Inception soundtrack
playlist = get_playlist("Inception", 2010)
if playlist:
    print("Test 1 — Inception soundtrack:")
    print(f"Name: {playlist['name']}")
    print(f"URL: {playlist['url']}")
    print(f"Image: {playlist['image']}")
else:
    print("Test 1 — No playlist found")
print()
```
