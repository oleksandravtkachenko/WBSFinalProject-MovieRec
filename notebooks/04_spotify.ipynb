```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from dotenv import load_dotenv
import os

load_dotenv()

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=os.getenv("SPOTIFY_ID"),
    client_secret=os.getenv("SPOTIFY_SECRET")
))

def get_playlist(mood, genres):
    query = f"{mood} {' '.join(genres[:2])} movie soundtrack"
    res = sp.search(q=query, type="playlist", limit=1)
    items = res["playlists"]["items"]
    if items:
        p = items[0]
        return {
            "name": p["name"],
            "url": p["external_urls"]["spotify"],
            "image": p["images"][0]["url"] if p["images"] else None
        }
    return None
```

```python
# Test 1 — happy mood
playlist = get_playlist("Happy", ["Comedy", "Romance"])
if playlist:
    print(f"Name: {playlist['name']}")
    print(f"URL: {playlist['url']}")
    print(f"Image: {playlist['image']}")
else:
    print("No playlist found")
```

```python
# Test 2 — tense mood
playlist = get_playlist("Tense", ["Thriller", "Action"])
if playlist:
    print("Test 2 — Tense Thriller:")
    print(f"Name: {playlist['name']}")
    print(f"URL: {playlist['url']}")
    print(f"Image: {playlist['image']}")
else:
    print("Test 2 — No playlist found")
print()
```

```python
# Test 3 — romantic mood
playlist = get_playlist("Romantic", ["Romance", "Drama"])
if playlist:
    print("Test 3 — Romantic Drama:")
    print(f"Name: {playlist['name']}")
    print(f"URL: {playlist['url']}")
    print(f"Image: {playlist['image']}")
else:
    print("Test 3 — No playlist found")
print()
```

```python
# Test 4 — check return type is dict with required keys
result = get_playlist("Happy", ["Comedy"])
if result:
    print(f"Test 4 — Returns dict: {isinstance(result, dict)}")
    print(f"Has required keys: {all(k in result for k in ['name', 'url', 'image'])}")
else:
    print("Test 4 — No playlist found — returned None")
```
