```python
import requests
# Follow Wikipedia's policy
HEADERS = {
    "User-Agent": "MovieRecAI/1.0 (student data science project)"
}

def get_summary(title, year=None):
    if year:
        slug = f"{title.replace(' ', '_')}_({year}_film)"
        r = requests.get(
            f"https://en.wikipedia.org/api/rest_v1/page/summary/{slug}",
            headers=HEADERS
        )
        if r.status_code == 200:
            return r.json().get("extract", "")[:300]

    # Fallback — (film) suffix
    slug = f"{title.replace(' ', '_')}_(film)"
    r = requests.get(
        f"https://en.wikipedia.org/api/rest_v1/page/summary/{slug}",
        headers=HEADERS
    )
    if r.status_code == 200:
        return r.json().get("extract", "")[:300]

    # Final fallback — just title
    slug = title.replace(' ', '_')
    r = requests.get(
        f"https://en.wikipedia.org/api/rest_v1/page/summary/{slug}",
        headers=HEADERS
    )
    return r.json().get("extract", "")[:300] if r.status_code == 200 else ""
```

```python
# Test 1 — well known movie with year
summary = get_summary("Inception", 2010)
print(f"Test 1 — Inception (2010):")
print(summary[:200])
print()
```

```python
# Test 2 — movie without year fallback
summary = get_summary("The Godfather", 1972)
print(f"Test 2 — The Godfather (1972):")
print(summary[:200])
print()
```

```python
# Test 3 — movie with year (tests year-specific Wikipedia slug)
summary = get_summary("Interstellar", 2014)
print(f"Test 3 — Interstellar (2014):")
print(summary[:200])
print()
```

```python
# Test 4 — non existent movie should return empty string
summary = get_summary("ThisMovieDoesNotExist", 2099)
print(f"Test 4 — Non existent movie:")
print(f"Returns empty string: {summary == ''}")
```
